```python
import matplotlib.pyplot as plt
import pandas as pd
# Sample sensor importance data
sensor_data = {
    'sensor': ['GPS', 'IMU', 'Battery'],
    'importance': [90, 85, 70]
}

sensor_df = pd.DataFrame(sensor_data)

# Plot sensor importance
plt.figure(figsize=(8, 5))
plt.bar(sensor_df['sensor'], sensor_df['importance'], color=['blue', 'orange', 'green'])
plt.title("Sensor Importance for Mission Performance", fontsize=14)
plt.xlabel("Sensor", fontsize=12)
plt.ylabel("Importance (%)", fontsize=12)
plt.ylim(0, 100)
plt.show()
```
