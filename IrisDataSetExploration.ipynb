# Simple Manual Model

```python
def single_feature_prediction (petal_length) :
    " " " Predict the Iris Species given the petal length based on decision tree classifiers " " "
    if petal_length < 2.5:
       return 0
    elif petal_length < 4.8:
       return 1
    else: 
       return 2
```

```python
df_train.columns
```

```python
x_train[:, 2]
```

```python
df_train['petal length (cm)'].values
```

```python
manual_y_predictions = np.array([single_feature_prediction(val) for val in x_train[:,2]])
```

```python
manual_model_accuracy = np.mean(manual_y_predictions == y_train)
```

```python
print(f"Manual Model Accuracy: {manual_model_accuracy * 100 : 2f}%")
```

# Modeling - Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Using Validation set to evaluate Model
```

```python
model = LogisticRegression()
```

```python
# Xt stands for X train and Xv stands for X validation
```

```python
Xt, Xv, Yt, Yv = train_test_split(x_train, y_train, test_size=0.25)
```

```python
Xt.shape
```

```python
Xv.shape
```

```python
model.fit(Xt, Yt) # you never want to evaluate  the model on the same data that was used for training
```

```python
y_pred = model.predict(Xv)
```

```python
np.mean(y_pred == Yv)
```

```python
model.score(Xv, Yv)
```

```python
from sklearn.metrics import accuracy_score,  confusion_matrix
print(accuracy_score(Yv, y_pred))
```

```python
confusionMatrix = confusion_matrix(Yv, y_pred)
print(confusionMatrix)
```

```python
ax = sns.heatmap(confusionMatrix, annot=True, fmt='d', cmap='Reds')
ax.set_title('Confusion Matrix for iris dataset prediction')
ax.set_xlabel('Prediction', fontsize=14)
ax.set_ylabel('Actual', fontsize=14)
```

# Using cross valiation for evaluation

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring="accuracy")
```
