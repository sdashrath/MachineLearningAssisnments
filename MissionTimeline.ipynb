```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample timestamp data
timeline_data = {
    'timestamp': [1, 2, 3, 4, 5],
    'event': ['Start', 'Waypoint 1', 'Waypoint 2', 'Waypoint 3', 'End']
}

timeline_df = pd.DataFrame(timeline_data)

# Plot mission timeline
plt.figure(figsize=(10, 4))
plt.scatter(timeline_df['timestamp'], [1] * len(timeline_df), color='red', label='Event')
for i, txt in enumerate(timeline_df['event']):
    plt.text(timeline_df['timestamp'][i], 1.1, txt, ha='center')

plt.title("Mission Timeline", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.yticks([])  # Hide y-axis as it's a timeline
plt.legend()
plt.grid(axis='x')
plt.show()
```
